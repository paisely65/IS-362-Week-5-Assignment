 ## What is the northernmost airport in the U.S.?

## What is the easternmost airport in the U.S.?
 To determine the easternmost airport the dataset was filtered to show airports with a latitude in the range of 44 to 50 and a longitude of -65 to -75 which would account for the airports on the far eastern corridor of the United States.
### The easternmost airport is Northern Aroostook Regional Airport, faa name WFK.

```python
df = pd.read_csv('https://raw.githubusercontent.com/maddogdatascience/DATA607/master/airports.csv')
northeast_airports = df[(df['lat'] > 44) & (df['lat'] <= 50) & (df['lon'] >= -75) & (df['lon'] <= -65)]
northeast_airports.sort_values('lat', ascending=False)
```

## On February 12th, 2013, which New York area airport had the windiest weather?


```python
# Performing a test of the data by viewing the first five rows.
speed = pd.read_csv("https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv")
speed.head()
```

```python
speed[(speed["month"]==2) & (speed["day"]==12) & (speed["year"]==2013)].sort_values("wind_speed", ascending=False)
```
